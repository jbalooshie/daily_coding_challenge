```python
# Write a program that will read a string from the user, and then it will print out how many times the following letters appear in the string: a, t, e, o, s
def string_reader():
    string = input('Please enter a string.').lower()
    letters = ['a', 't', 'e', 'o', 's']
    letters_dict = {}
    for x in letters:
        counter = 0
        for y in string:
            if x == y:
                counter += 1
        letters_dict.update({x:counter})
    return letters_dict
                

print(string_reader())
```

```python
# Write a Python program that will read a number from the user, and then it will check and print if that number is a prime number or not. 
# A prime number is the one that is divided exactly only with itself and 1 (e.g. 2, 3, 5, 7, 11, ....). 
# For example, for number 7 it should print "7 is a prime number" while for 8 it should print "8 is not a prime number". 
# Keep in mind that too big numbers could hang your program running for a long time, so restrict your tests to small numbers (< 1000).
from math import sqrt
def prime_checker():
    n = int(input('Please input a number to check'))
    if (n <= 1):
        return f'{n} is not a prime number'
    for i in range(2, int(sqrt(n))+1):
        if (n % i == 0):
            return f'{n} is not a prime number'

    return f'{n} is a prime number'
        
    


prime_checker()

```

```python
# Aaron and Alex found a game online that asks for a secret password to enter. This password is a 3-digit number. 
# They also know that the first digit is greater than the second one and that the third digit is equal to the sum of the first and second digit.
# Write a Python program that will print out all possible combinations of the secret password.
import time
start_time = time.time()
def secret_password():
    passwords = []
    for x in range(1,9):
        for y in range(0,x):
            z = x + y
            if z >= 10:
                continue
            else:
                passwords.append(str(x) + str(y) + str(z))
            
    return print(passwords)

secret_password()
print('My program took', time.time() - start_time, 'to run')
```

```python
# Write a Python program that will read first a number N, and then it will read N numbers from the user, and it will count how many positive and negative numbers were given.
def int_counter():
    n = int(input('Please input a number'))
    provided_nums = []
    positive_nums = 0
    negative_nums = 0
    for x in range(0, n):
        user_num = int(input(f'Please input an additional number. You will be asked to provide {n} additional number(s).'))
        provided_nums.append(user_num)
    for x in provided_nums:
        if x > 0:
            positive_nums += 1
        if x < 0:
            negative_nums += 1
        if x == 0:
            continue
    return print(f'You provided {positive_nums} positive numbers and {negative_nums} negative numbers.')

int_counter()
                             
        
```

```python
# An advert on the PythonTV is incrementally calculated according to the following. 
# The first 15 minutes 50 EUR per second, from 16 to 30 minutes 30 EUR per second, from 31 minutes and beyond 20 EUR per second. 
# Create a Python program that will calculate the cost of an advert on the PythonTV given the duration of the advert in minutes. 
# For example, for an advert of 31 minutes, the cost should be 73,200 EUR while for 1 minute it should cost 3,000 EUR.

def ad_cost():
    length = int(input('Please enter the total length of your ad in seconds. For example, if your ad is 2 minutes, 30 seconds long, please enter 150.'))
    if length <= 900:
        cost = length * 50
        return print(f'Your advert of {length} seconds will cost {cost} EUR to air.')
    elif length > 900 and length <= 1800:
        cost = ((length - 900) * 30) + 45000
        return print(f'Your advert of {length} seconds will cost {cost} EUR to air.')
    else:
        cost = ((length - 1800) * 20) + 72000
        return print(f'Your advert of {length} seconds will cost {cost} EUR to air.')

ad_cost()
```

```python
# Write a Python program that will calculate the sum of the following sequence 1^2 + 2^2 + 3^2 + 4^2 + … + n^2 for a given n number.
def sequence_sum(n):
    if n == 0 or n == 1:
        return n ** 2
    else:
        result = 0
        for x in range(1,n+1):
            squared = x ** 2
            result += squared
        return result

print(sequence_sum(25))
```

```python
# A student from this course asked his parents to buy him a faster laptop, so he can start doing professional work with his programming skills in Python. 
# The laptop costs £1500. His parents, can’t afford to pay this amount but instead, they can give him double the money they gave him the week before, every week. 
# They said they will start with £45. 
# Create a program in Python that will compute in how many weeks the student will collect enough money to buy the laptop, and what the amount he will collect by then looks like.

def laptop_cost(starting_amount, laptop_cost):
    previous_week = starting_amount
    total_accrued = starting_amount
    weeks_counter = 0
    while total_accrued < laptop_cost:
        previous_week = previous_week * 2
        total_accrued = total_accrued + previous_week
        weeks_counter = weeks_counter + 1
    print(f'It will take {weeks_counter} weeks to accrue ${total_accrued} to buy the laptop.')

laptop_cost(45,1500)

```

```python
# Write a Python program that will read an integer number from the user representing seconds, and it will print the seconds in terms of days, hours, minutes, and seconds. 
# For example, if the input is 125 seconds it should print 2 minutes and 5 seconds. If the input, however, is 172805 then it should print 2 days and 5 seconds etc.

def convert_seconds(total_seconds: int):
    SECONDS_PER_MINUTE = 60
    SECONDS_PER_HOUR = 3600
    SECONDS_PER_DAY = 86400
    days = total_seconds // SECONDS_PER_DAY
    remaining_seconds = total_seconds % SECONDS_PER_DAY
    hours = remaining_seconds // SECONDS_PER_HOUR
    remaining_seconds %= SECONDS_PER_HOUR
    minutes = remaining_seconds // SECONDS_PER_MINUTE
    seconds = remaining_seconds % SECONDS_PER_MINUTE
    result = []
    if days > 0:
        result.append(f"{days} day{'s' if days > 1 else ''}")
    if hours > 0:
        result.append(f"{hours} hour{'s' if hours > 1 else ''}")
    if minutes > 0:
        result.append(f"{minutes} minute{'s' if minutes > 1 else ''}")
    if seconds > 0:
        result.append(f"{seconds} second{'s' if seconds > 1 else ''}")

    print(" and ".join(result))

convert_seconds(4000)
```

```python
# Write a function in Python that when a password is given as a string, it will return True if the password is strong, False otherwise.
# A password is strong if it has at least 8 characters, it contains at least a number and a symbol, and contains both an uppercase and a lowercase letter.
import re
def password_strength(password):
    if len(password) < 8:
        return False
    if not re.search(r'[A-Z]', password):
        return False
    if not re.search(r'[a-z]', password):
        return False
    if not re.search(r'\d', password):
        return False
    if not re.search(r'[!@#$%^&*(),.?":{}|<>]', password):
        return False
    return True
print(password_strength('!Python1'))
print(password_strength('Python1'))
```

```python
# Write a function in Python that will generate a random strong password according to the definition of a strong password from the previous exercise. 
# The function will accept a single parameter, the length of the desired strong password. You might need to use the random library. 
# If you import random and then use random.choice("abcdefghijklmnop....") it will give you a random character from the string "abcdefghijklmnop....". 
# Use this library and the random.choice() function to generate a random strong password of the length of the user (given as an argument). 
# The password doesn't have to mean something, it could be just a random sequence of letters that meet the requirements of a strong password.
import random

def password_generator(password_length):
    lower_letters = 'abcdefghijklmnopqrstuvwxyz'
    capital_letters = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    numbers = '123456789'
    symbols = '[!@#$%^&*(),.?":{}|<>]'
    password = ''
    if password_length < 8:
        return('Password must by 8 or more characters.')
    for x in range(password_length-3):
        letter = random.choice(lower_letters)
        password = password + letter
    password = password + random.choice(capital_letters) + random.choice(numbers) + random.choice(symbols)
    password = ''.join(random.sample(password,len(password)))
    return password

print(password_generator(20))
```

```python
# Write a Python function that will approximate π. The function will take a single argument, the approximation length of π and the function should generate the π. 
# The formula is given by π = 3 + (4 / (2 * 3 * 4)) - (4 / (4 * 5 * 6)) + (4 / (6 * 7 * 8)) - (4 / (8 * 9 * 10)) + ... )
from math import acos
def pi_slice(slice_length):
    result = round(2 * acos(0.0), slice_length)
    return result


pi_slice(10)
```
